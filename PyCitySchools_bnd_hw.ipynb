```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
school_file = "schools_complete.csv"
student_file = "students_complete.csv"

# Read School and Student Data File and store into Pandas Data Frames
school_df = pd.read_csv(school_file)
# print(school_df.columns)

student_df = pd.read_csv(student_file)
# print(student_df.columns)

# Combine the data into a single dataset
merged_school_df = pd.merge(student_df, school_df, how="left", on=["school_name", "school_name"])
merged_school_df.head()
```

## District Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average math score 

* Calculate the average reading score

* Calculate the overall passing rate (overall average score), i.e. (avg. math score + avg. reading score)/2

* Calculate the percentage of students with a passing math score (70 or greater)

* Calculate the percentage of students with a passing reading score (70 or greater)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
#Calculate the total number of schools
total_num_schools = len(merged_school_df['school_name'].unique().tolist())
# print(total_num_schools)

#Calculate the total number of students
total_num_students = merged_school_df['student_name'].count()
# print(total_num_students)

#Calculate the total budget
# total_school_budget = merged_school_df.groupby(['school_name'])['budget'].agg('sum')
total_school_budget = school_df['budget'].sum()
# print(total_school_budget)

#Calculate the average math score
avg_math_score = merged_school_df['math_score'].mean()
# print(avg_math_score)

#Calculate the average reading score
avg_reading_score = merged_school_df['reading_score'].mean()
# print(avg_reading_score)

#Calculate the overall passing rate (overall average score), i.e. (avg. math score + avg. reading score)/2
overall_avg_score = (avg_math_score + avg_reading_score)/2
# print(overall_avg_score)

#Calculate the percentage of students with a passing math score (70 or greater)
passing_math_students = merged_school_df.loc[merged_school_df['math_score'] >= 70]['math_score'].count()
pct_passing_math = passing_math_students/total_num_students * 100
# print(pct_passing_math)

#Calculate the percentage of students with a passing reading score (70 or greater)
passing_reading_students = merged_school_df.loc[merged_school_df['reading_score'] >= 70]['reading_score'].count()
pct_passing_reading = passing_reading_students/total_num_students * 100
# print(pct_passing_reading)

#Create a dataframe to hold the above results
district_summary_df = pd.DataFrame({
    "Total Schools" : [total_num_schools],
    "Total Students": [total_num_students],
    "Total Budget" : [total_school_budget],
    "Avg. Math Score" : [avg_math_score],
    "Avg. Reading Score" : [avg_reading_score],
    "% Passing Math" : [pct_passing_math],
    "% Passing Reading" : [pct_passing_reading],
    "% Overall Passing Rate" : [overall_avg_score]  
}) 
district_summary_df.head()
```

## Scores by School Spending

* Create a table that breaks down school performances based on average Spending Ranges (Per Student). Use 4 reasonable bins to group school spending. Include in the table each of the following:
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)

```python
# Sample bins. Feel free to create your own bins.
spending_bins = [0, 585, 615, 645, 675]
group_names = ["<$585", "$585-615", "$615-645", "$645-675"]
```

```python
merged_school_df['spending_bins'] = pd.cut(merged_school_df['budget']/merged_school_df['size'], spending_bins, labels = group_names)

#group by spending
by_spending = merged_school_df.groupby('spending_bins')

#calculations
avg_math = by_spending['math_score'].mean()
avg_read = by_spending['reading_score'].mean()
pass_math = merged_school_df[merged_school_df['math_score'] >= 70].groupby('spending_bins')['Student ID'].count()/by_spending['Student ID'].count() *100
pass_read = merged_school_df[merged_school_df['reading_score'] >= 70].groupby('spending_bins')['Student ID'].count()/by_spending['Student ID'].count() *100
overall = merged_school_df[(merged_school_df['reading_score'] >= 70) & (merged_school_df['math_score'] >= 70)].groupby('spending_bins')['Student ID'].count()/by_spending['Student ID'].count() *100

            
# df build            
scores_by_spend = pd.DataFrame({
    "Average Math Score": [avg_math],
    "Average Reading Score": [avg_read],
    '% Passing Math': [pass_math],
    '% Passing Reading': [pass_read],
    "Overall Passing Rate": [overall]
            
})
            
#reorder columns
scores_by_spend = scores_by_spend[[
    "Average Math Score",
    "Average Reading Score",
    '% Passing Math',
    '% Passing Reading',
    "Overall Passing Rate"
]]

scores_by_spend.index.name = "Per Student Budget"
scores_by_spend = scores_by_spend.reindex(group_names)

scores_by_spend.head()
```
